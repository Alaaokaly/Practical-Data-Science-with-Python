```python
import pandas as pd

df = pd.read_excel('data/HIV_results.xlsx')
df
```

```python
df.drop('Name', axis=1, inplace=True)
df.loc[(df['Age'] >= 20) & (df['Age'] < 30), 'Age'] = 20
df.loc[(df['Age'] >= 30) & (df['Age'] < 40), 'Age'] = 30
df.loc[(df['Age'] >= 40) & (df['Age'] < 50), 'Age'] = 40
df['Age'] = df['Age'].apply(lambda x: str(x) + 's')
df['Zipcode'] = df['Zipcode'].apply(lambda x: str(x)[:3] + '**')
```

```python
df
```

```python
df['equiv_class'] = [0, 1, 0, 2, 1, 2]
```

```python
df.sort_values(by='equiv_class')
```
