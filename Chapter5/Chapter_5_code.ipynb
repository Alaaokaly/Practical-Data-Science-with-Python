### Choosing the right method for plotting

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///data/chinook.db')

with engine.connect() as connection:
    sql_df = pd.read_sql_table('invoices', connection)
```

```python
sql_df.head()
```

```python
sql_df.groupby('BillingCountry').sum().sort_values(by='Total', ascending=False)[:3]
```

```python
top_3_countries = sql_df.groupby('BillingCountry').sum().sort_values(by='Total', ascending=False)[:3].index.values
```

```python
top_3_countries
```

```python
sql_df.set_index('InvoiceDate', inplace=True)
```

```python
gb = sql_df[sql_df['BillingCountry'].isin(top_3_countries)]. \
        groupby([pd.Grouper(freq='M'), 'BillingCountry']).sum(). \
        groupby(level=-1).cumsum()
```

```python
gb
```

```python
gb.reset_index(inplace=True)
```

```python
gb.head()
```

```python
sns.lineplot(data=gb, x='InvoiceDate', y='Total', hue='BillingCountry')
```

```python
# saving figure for book
sns.lineplot(data=gb, x='InvoiceDate', y='Total', hue='BillingCountry')
plt.tight_layout()
plt.savefig('seaborn_lineplot_color.png', dpi=300)
```

```python
# black and white redundancy
sns.lineplot(data=gb, x='InvoiceDate', y='Total', hue='BillingCountry', style='BillingCountry')
```

```python
# save figure for book
sns.lineplot(data=gb, x='InvoiceDate', y='Total', hue='BillingCountry', style='BillingCountry')
plt.tight_layout()
plt.savefig('seaborn_lineplot_color_dashes.png', dpi=300)
```

```python
sns.lineplot(data=gb, x='InvoiceDate', y='Total', hue='BillingCountry', style='BillingCountry', dashes=[(2, 1), (5, 2), ''])
```

```python
from labellines import labelLines
```

```python
f = plt.figure()
ax = f.gca()
for country in top_3_countries:
    c_df = gb[gb['BillingCountry'] == country]
    ax.plot(c_df['InvoiceDate'], c_df['Total'], label=country)

labelLines(ax.get_lines())

plt.xlabel('Year')
plt.ylabel('Cumulative Sales ($)')

# save image for book
plt.tight_layout()
plt.savefig('seaborn_lineplot_labels.png', dpi=300)
```

### Another way to collect and plot the same data

```python
gb = sql_df[sql_df['BillingCountry'].isin(top_3_countries)].groupby([pd.Grouper(freq='M'), 'BillingCountry']).sum()
dfs = []
for country in top_3_countries:
    c_df = gb.xs(country, level=1).cumsum()
    c_df['Country'] = country
    dfs.append(c_df)
    
full_c_df = pd.concat(dfs)
```

```python
full_c_df.head()
```

```python
full_c_df.reset_index(inplace=True)
```

```python
sns.lineplot(data=full_c_df, x='InvoiceDate', y='Total', hue='Country')
```

### Here is a start how we could collect the data using a SQL query.
We would need to either modify the SQL query to join the same data for the top 3 countries, or do 3 separate SQL queries and join the resulting dataframes.

```python
engine = create_engine('sqlite:///data/chinook.db')

query = """SELECT SUM(Total) OVER (ORDER BY InvoiceId) as Total, strftime("%m-%Y", InvoiceDate) as 'month-year', BillingCountry
FROM invoices
WHERE BillingCountry="USA"
GROUP BY strftime("%m-%Y", InvoiceDate);
"""

with engine.connect() as connection:
    sql_df2 = pd.read_sql_query(query, connection)
```

```python
sql_df2.head()
```

```python
sns.lineplot(data=sql_df2, x='month-year', y='Total', hue='BillingCountry')
```
